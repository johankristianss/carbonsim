```python
import matplotlib.pyplot as plt
import pandas as pd

#logidf = pd.read_csv('/scratch/cognit/carbon/PL.csv')
df = pd.read_csv('/scratch/cognit/carbon/PL.csv')
#
#df = pd.read_csv('/scratch/cognit/carbon/SE-SE4.csv')
df = pd.read_csv('/scratch/cognit/carbon/SE-SE4.csv')
df['datetime'] = pd.to_datetime(df['datetime'])

plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['carbonIntensity'], linestyle='-', color='b')

plt.title('Carbon Intensity Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Carbon Intensity (gCOeq/kWh)')
plt.ylim(0, 1100)  
#plt.ylim(0, 100)  
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
