```python
import matplotlib.pyplot as plt
import pandas as pd

#logidf = pd.read_csv('/scratch/cognit/carbon/PL.csv')
#df = pd.read_csv('/scratch/cognit/carbon/FR.csv')
#
#df = pd.read_csv('/scratch/cognit/carbon/SE-SE4.csv')
df = pd.read_csv('/scratch/cognit/carbon/SE-SE1.csv')
df = pd.read_csv('/scratch/cognit/carbon/GB.csv')
df['datetime'] = pd.to_datetime(df['datetime'])

start_time = df['datetime'].min()
end_time = start_time + pd.Timedelta(hours=24)
df = df[df['datetime'] < end_time]

plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['carbonIntensity'], linestyle='-', color='b')

plt.title('Carbon Intensity Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Carbon Intensity (gCOeq/kWh)')
#plt.ylim(0, 20)  

plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
