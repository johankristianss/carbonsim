```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import matplotlib.dates as mdates

df = pd.read_csv('/scratch/cognit/carbon/PL.csv')
#df = pd.read_csv('/scratch/cognit/carbon/FR.csv')
#
#df = pd.read_csv('/scratch/cognit/carbon/SE-SE3.csv')
df = pd.read_csv('/scratch/cognit/carbon/SE-SE1.csv')
df = pd.read_csv('/scratch/cognit/carbon/GB.csv')
df['datetime'] = pd.to_datetime(df['datetime'])

start_time = df['datetime'].min()
end_time = start_time + pd.Timedelta(hours=100)
df = df[df['datetime'] < end_time]

plt.figure(figsize=(10, 6))
plt.plot(df['datetime'], df['carbonIntensity'], linestyle='-', color='g')

plt.title('Warsaw', fontsize=24, fontweight='bold')
plt.xlabel('Date and Time', fontsize=22, fontweight='bold')
plt.ylabel('CO2 Intensity (gCOeq/kWh)', fontsize=22, fontweight='bold')
#plt.ylim(0, 20)  

plt.grid(True)
plt.xticks(rotation=45)

#ax = plt.gca()
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))  # '%b' for abbreviated month names
#ax.xaxis.set_major_locator(mdates.MonthLocator()) 

output_image_path = os.path.join("./warsaw.png")
plt.savefig(output_image_path, dpi=300)  # You can adjust dpi as needed


plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```
