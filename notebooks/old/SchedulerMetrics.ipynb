```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('/scratch/cognit/scheduler_metrics_best.csv')

# Convert ticks to hours
df['hours'] = df['tick'] / 3600

# Plot the cumulative emission over time (hours)
plt.figure(figsize=(10, 6))
plt.plot(df['hours'], df['cumulative_emission'], label='Cumulative Emission [g]', color='red')
plt.xlabel('Time [hours]')
plt.ylabel('Cumulative Emission [g]')
plt.title('Cumulative Emission Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot the cumulative energy over time (hours)
plt.figure(figsize=(10, 6))
plt.plot(df['hours'], df['cumulative_energy'], label='Cumulative Energy [kWh]', color='blue')
plt.xlabel('Time [hours]')
plt.ylabel('Cumulative Energy [kWh]')
plt.title('Cumulative Energy Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
