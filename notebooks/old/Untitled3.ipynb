```python
import os
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Directories
input_dir = "../filtered_workloads_1s_stats_max_waittime"

# List CSV files in the directory
csv_files = [os.path.join(input_dir, f) for f in os.listdir(input_dir) if f.endswith(".csv")]

# Initialize a list to store workload summaries
workload_data = []

# Read and aggregate data from all CSV files
for file in csv_files:
    df = pd.read_csv(file)
    # Extract summary metrics for the workload
    workload_summary = {
        "workload": os.path.basename(file),  # File name as workload identifier
        "mean_power_draw": df.loc[df["column_name"] == "power_draw_W", "mean"].values[0],
        "mean_gpu_utilization": df.loc[df["column_name"] == "utilization_gpu_pct", "mean"].values[0],
        "mean_memory_utilization": df.loc[df["column_name"] == "utilization_memory_pct", "mean"].values[0],
        "total_length_seconds": df.loc[df["column_name"] == "power_draw_W", "total_length_seconds"].values[0],
        "max_waittime": df.loc[df["column_name"] == "power_draw_W", "max_waittime"].values[0],
    }
    workload_data.append(workload_summary)

# Convert to DataFrame
workloads_df = pd.DataFrame(workload_data)

# Select features for clustering
features = ["mean_power_draw", "mean_gpu_utilization", "mean_memory_utilization", "total_length_seconds", "max_waittime"]

# Standardize features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(workloads_df[features])

# Perform k-means clustering
optimal_k = 3  # Adjust this based on experimentation (e.g., elbow method)
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
workloads_df["cluster"] = kmeans.fit_predict(scaled_features)

# Reduce dimensions for visualization (PCA)
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(scaled_features)

# Plot clusters
plt.figure(figsize=(10, 7))
for cluster in range(optimal_k):
    cluster_points = reduced_features[workloads_df["cluster"] == cluster]
    plt.scatter(
        cluster_points[:, 0],
        cluster_points[:, 1],
        label=f"Cluster {cluster}",
        alpha=0.7
    )

plt.title("Clusters of Workloads")
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.legend()
plt.show()

# Display cluster summaries
print(workloads_df.groupby("cluster").mean())
```
