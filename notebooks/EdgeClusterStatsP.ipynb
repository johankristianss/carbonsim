```python
import pandas as pd
import matplotlib.pyplot as plt

#dir = "../results/30/genetic_timepool"
dir = "../results/30/greedy_binpack"
edge_cluster1 = "Lulea"
edge_cluster2 = "Warsaw"

# Load the data
df1 = pd.read_csv(dir + '/' + edge_cluster1 + '.csv')
df1['hours'] = df1['tick'] / 3600
df1['cluster_cumulative_emission'] = df1['cluster_cumulative_emission']

df2 = pd.read_csv(dir + '/' + edge_cluster1 + '.csv')
df2['hours'] = df2['tick'] / 3600
df2['cluster_resource_utilization'] = df2['cluster_resource_utilization'] * 100

df3 = pd.read_csv(dir + '/'+ edge_cluster2 + '.csv')
df3['hours'] = df3['tick'] / 3600
df3['cluster_cumulative_emission'] = df3['cluster_cumulative_emission']

df4 = pd.read_csv(dir + '/' + edge_cluster2 + '.csv')
df4['hours'] = df4['tick'] / 3600
df4['cluster_resource_utilization'] = df4['cluster_resource_utilization'] * 100

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot cumulative emissions on the primary y-axis
ax1.plot(df1['hours'], df1['cluster_cumulative_emission'], label=edge_cluster1 + ' - Emission', color='blue')
ax1.plot(df3['hours'], df3['cluster_cumulative_emission'], label=edge_cluster2 + ' - Emission', color='green')
ax1.set_xlabel('Time [hours]')
ax1.set_ylabel('Cumulative Emission [g]', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Add the secondary y-axis for utilization
ax2 = ax1.twinx()
ax2.plot(df2['hours'], df2['cluster_resource_utilization'], label=edge_cluster1 + ' - Utilization', color='blue', linestyle='dashed')
ax2.plot(df4['hours'], df4['cluster_resource_utilization'], label=edge_cluster2 + ' - Utilization', color='green', linestyle='dashed')
ax2.set_ylabel('Resource Utilization [%]', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Add legends for both axes
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
ax1.grid(True)

# Add a title
plt.title('CO2 Emissions and Resource Utilization Over Time')

# Show the plot
plt.tight_layout()
plt.show()
```
