```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the exponential distribution
rate = 1  # Rate parameter
scale = 1  # Scale constant

# Generate random samples from the exponential distribution
samples = np.random.exponential(1, size=10000) * 100

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=50, density=True, alpha=0.75, color='skyblue', label='Sampled Distribution')

# Plot the theoretical distribution
#x = np.linspace(0, np.max(samples), 500)
#pdf = rate * np.exp(-rate * x / scale)
#plt.plot(x, pdf, 'r-', label=f'Theoretical PDF\nRate: {rate}, Scale: {scale}')

# Add titles and labels
plt.title('Exponential Distribution', fontsize=16)
plt.xlabel('Waiting Time', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
